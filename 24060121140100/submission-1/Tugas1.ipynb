```python
import pandas as pd
```

```python
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
```

```python
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
```

```python
import matplotlib.pyplot as plt

selected_columns = ['age', 'sex', 'chol', 'thalach']

# Membuat plot univariat dengan jenis plot box untuk kolom yang dipilih
heart_disease_dataset[selected_columns].plot(kind='box', subplots=True, layout=(2, 2), sharex=False, sharey=False)
plt.show()
```

Boxplot adalah alat yang berguna untuk menggambarkan distribusi data dalam bentuk grafik yang memperlihatkan nilai-nilai statistik penting seperti median, kuartil, dan adanya pencilan (outliers) :


* Kotak (box): Kotak pada boxplot menunjukkan jangkauan antara kuartil pertama (Q1) dan kuartil ketiga (Q3), yang mengandung sebagian besar data (interquartile range, IQR).
*   Garis tengah dalam kotak: Ini menunjukkan median (nilai tengah) dari data.


*   Garis-garis bawah dan atas (whiskers): Garis-garis ini menunjukkan rentang data di luar kotak (di luar IQR). Garis bawah mencakup data terendah sedangkan garis atas mencakup data tertinggi.
*   Pencilan (outliers): Titik-titik di luar garis whiskers adalah pencilan atau nilai yang dianggap tidak biasa dalam distribusi data.
